- #### Categorising the age, date and year in standard format

- #### mean & meadian values should not be in negative, converting in correct format

```python
main_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION", "DAYS_ID_PUBLISH", "DAYS_LAST_PHONE_CHANGE"]]=abs (main_data[["DAYS_BIRTH","DAYS_EMPLOYED","DAYS_REGISTRATION", "DAYS_ID_PUBLISH", "DAYS_LAST_PHONE_CHANGE"]])
```

```python
#### creating age groups
main_data["AGE"] = main_data["DAYS_BIRTH"]/365
bins = [0,20,25,30,35,40,45,50,55,60,100]
slots = ["0-20","20-25","25-30","30-35","35-40","40-45","45-50","50-55","55-60","60 Above"]
main_data["AGE_GROUP"] = pd.cut(main_data["AGE"], bins=bins, labels=slots)
main_data["AGE_GROUP"].value_counts(normalize= True)*100
```

```python
#### Converting employment days to employment year for more redability
main_data["YEARS_EMPLOYED"] = main_data["DAYS_EMPLOYED"]/365
bins = [0,5,10,15,20,25,30,50]
slots = ["0-5","5-10","10-15","15-20","20-25","25-30","30 Above"]

main_data["EMPLOYEMENT_YEARS"] = pd.cut(main_data["YEARS_EMPLOYED"], bins=bins, labels=slots)
main_data["EMPLOYEMENT_YEARS"].value_counts(normalize= True)*100
```

```python
#### Converting registration days to registration year for more redability
main_data["YEARS_REGISTRATION"] = main_data["DAYS_REGISTRATION"]/365
bins = [0,5,10,15,20,25,30,50]
slots = ["0-5","5-10","10-15","15-20","20-25","25-30","30 Above"]

main_data["REGISTRATION_YEARS"] = pd.cut(main_data["YEARS_REGISTRATION"], bins=bins, labels=slots)
main_data["REGISTRATION_YEARS"].value_counts(normalize= True)*100
```

```python
corr_repayer = Repayer_data.corr()
corr_df_repayer = corr_repayer.where(np.triu(np.ones(corr_repayer.shape),k=1).astype(np.bool)).unstack().reset_index()
corr_df_repayer.columns =['VAR1','VAR2','Correlation']
corr_df_repayer.dropna(subset = ["Correlation"], inplace = True)
corr_df_repayer["Correlation"]=corr_df_repayer["Correlation"].abs() 
corr_df_repayer.sort_values(by='Correlation', ascending=False, inplace=True) 
corr_df_repayer.head(10)
```

```python
fig = plt.figure(figsize=(25,15))
ax = sns.heatmap(Repayer_data.corr(), cmap="RdYlGn",annot=True,linewidth =1)
plt.title("Heat map for Repayers")
plt.show()
```

##### Total Income
#### 2. We can also see that repayers have high correlation in number of days employed.
